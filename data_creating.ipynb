```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
DataDir= 'D:\\CarBike\\train'
CATEGORIES=["Bike","Car"]

for i in CATEGORIES:
    path=os.path.join(DataDir,i)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array,cmap='gray')
        plt.show()
        break
    break
```

```python
img_size=100

new_array=cv2.resize(img_array,(img_size,img_size))
plt.imshow(new_array,cmap='gray')
plt.show()
```

```python
training_data=[]

def create_training_data():
    for i in CATEGORIES:

        path=os.path.join(DataDir,i)
        class_num=CATEGORIES.index(i)

        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(img_size,img_size))
                training_data.append([new_array,class_num])
            
            except Exception as e:
                pass


create_training_data()
```

```python
print(len(training_data))
```

```python
print(len(training_data)) #lenth of our training set

#shuffle the data 
import random 
random.shuffle(training_data)
for sample in training_data[:10]:
    print(sample[1])
```

```python
b=[]
c=[]
```

```python
for features,label in training_data:
    b.append(features)
    c.append(label)
b=np.array(b).reshape(-1,img_size,img_size,1)
c=np.array(c)
```

```python
#saving it
import pickle 

pickle_out=open("b.pickle","wb")
pickle.dump(b,pickle_out)
pickle_out.close()

pickle_out=open("c.pickle","wb")
pickle.dump(c,pickle_out)
pickle_out.close()
```

```python
# to read it 
pickle_in=open("b.pickle","rb")
b=pickle.load(pickle_in)
```
